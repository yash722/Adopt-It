```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import numpy as np
import time
import urllib.request

opener = urllib.request.build_opener()
opener.addheaders = [('User-agent', 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36')]
urllib.request.install_opener(opener)
```

```python
# Run once (Already Ran)!
options = webdriver.ChromeOptions()
options.add_argument('--headless')
browser = webdriver.Chrome(options=options)
browser.get("https://www.mspca.org/adoption-search/?type=dog")
time.sleep(2)
dogs = browser.find_elements(By.CLASS_NAME, 'petCard')
data = {
    "Name" : [],
    "Breed":[],
    "Location": [],
    "Gender" : [],
    "Age":[],
    "Weight":[],
    "Image": [],
    "Description":[]
}
dogs = [i for i in dogs if i.text != ""]
l = ["Location", "Gender", "Age", "Weight"]
for dog in dogs:
    dog_link = dog.get_attribute("href")
    print(dog_link)
    time.sleep(5)
    br_dog = webdriver.Chrome(options=options)
    br_dog.get(dog_link)
    dog_image = br_dog.find_element(By.ID, "extraimage0")
    dog_image = dog_image.get_attribute("src")
    data["Image"].append(dog_image)
    dog_name = br_dog.find_element(By.TAG_NAME, "h1").text
    data["Name"].append(dog_name)
    dog_breed = br_dog.find_element(By.XPATH, '/html/body/div[6]/div[1]/div[1]/div[2]/h2').text
    data["Breed"].append(dog_breed)
    print(dog_image, dog_name, dog_breed)
    dog_info = br_dog.find_elements(By.CLASS_NAME, "petStatContent")
    dog_info = [i for i in dog_info if i.text != ""]
    for i in range(len(dog_info)):
        dog_data = dog_info[i].find_element(By.TAG_NAME, "span").text
        data[l[i]].append(dog_data)
        print(dog_data)
    dog_bio = br_dog.find_element(By.CLASS_NAME, "petBio")
    dog_desc = dog_bio.find_elements(By.TAG_NAME, "p")
    total_description = ""
    for desc in dog_desc:
        total_description += " " + desc.text
    data["Description"].append(total_description)
    br_dog.close()


browser.close()
```

```python
print(data)
```

```python
df_dogs = pd.DataFrame(data)
```
